```python
Profit_plot = (ggplot(sample, aes(x="Sub-Category" , y="Profit" , fill="Sub-Category")) + geom_col() + coord_flip() 
+ scale_fill_brewer(type="div" , palette="Spectral") + theme_classic() + ggtitle("Pie Chart"))

display(Profit_plot)
```

The Above Pie Chart  shows the Profit and Loss of each and every Sub-Categories

Here from the graph we can visulaise that "Binders" sub category has sufffered the highest amount of loss and also profit amonst all other sub categories
(For now we cants say the reason it may be because of discounts given on binders subcategory)

Next "Copiers" sub category has gained highest amount of profit with no loss. There are other sub categories too havent faced any kind of losses but their profit margins are also low.

Next, Suffering from highest loss is "Machines"

1st we ll check sales from every segments of whole data

```python
flip_xlabels = theme(axis_text_x = element_text(angle = 90, vjust = 0.5, hjust=1),figure_size=(10,5),
                    axis_ticks_length_major =10 ,axis_ticks_length_minor =5 )
(ggplot(sample, aes(x="Sub-Category" , fill="Sales")) + geom_bar() + facet_wrap(["Segment"])
 + flip_xlabels + ggtitle("Sales from every Segment of United States of whole Data"))
```

```python
# Analyzing Category wise sales, Discount, Profit:


category_data = sample1.groupby(['Category'])[['Sales', 'Discount', 'Profit']].mean()
category_data
```

```python
# Pie Charts of Category wise sales, Discount, Profit
category_data.plot.pie(subplots=True, figsize=(20, 18),autopct='%0.1f%%', labels = category_data.index)
plt.show()
```

Inference:

1)Technology Category has the maximum Sales and Profit.

2)Office Supplies category has minimum sales and Furniture category has lowest Profit.
